```python
import pandas as pd
weather = pd.read_csv("mandya dataset.csv")
weather
```

```python
weather.apply(pd.isnull).sum()/weather.shape[0]
```

```python
core_column=["YEAR", "DOY", "T2M", "T2MDEW", "T2M_MAX", "T2M_MIN", "RH2M", "PRECTOTCORR", "WS2M", "WS2M_MAX", "WS2M_MIN", "Place"]
core_weather = weather[core_column]
```

```python
core_weather
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#visulizationn of features
features_column = ["T2M", "T2MDEW", "T2M_MAX", "T2M_MIN", "RH2M", "PRECTOTCORR", "WS2M", "WS2M_MAX", "WS2M_MIN"]
plt.figure(figsize=(15, 10))
core_weather[features_column].hist(bins=20, range=(0, 100), color='blue', alpha=0.7, edgecolor='black')
plt.tight_layout()
plt.show()
#correlation b/w feeatures
correlation_matrix = core_weather[features_column].corr()
plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation of features")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_featuers = scaler.fit_transform(core_weather[features_column])
core_weather_scaled = pd.DataFrame(scaled_featuers, columns=features_column)

#spliting data into training and testing set
x = core_weather_scaled.drop(["PRECTOTCORR"],axis=1)
y = core_weather_scaled[["PRECTOTCORR"]]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)

print("shape of x_train:", x_train.shape)
print("shape of x_test:", x_test.shape)
print("shape of y_train:", y_train.shape)
print("shape of y_test:", y_test.shape)
```
